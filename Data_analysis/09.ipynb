## 나이와 월급의 관계

### 나이와 월급의 관계 분석

```python
# 나이별 월급 평균표
age_income = welfare.dropna(subset = ['income']).groupby('age').agg(mean_income = ('income', 'mean'))
age_income.head()
```

```python
# 선그래프
sns.lineplot(data = age_income, x = 'age', y = 'mean_income')
```

파생변수 만들기 - 연령대 (초년층 : 30세 미만, 중년층 : 30 ~ 59세, 노년층 : 60세 이상)

```python
welfare['age'].head()
```

```python
#연령대 변수 만들기
welfare = welfare.assign(ageg = np.where(welfare['age'] < 30, 'young',
                                         np.where(welfare['age'] <= 59, 'middle',
                                                  'old')))
welfare['ageg'].value_counts()
```

```python
# 빈도 막대 그래프
sns.countplot(data = welfare, x = 'ageg')
```

연령대별 월급 차이 분석

```python
# 연령대별 월급 평균표
ageg_income = welfare.dropna(subset = ['income'])\
.groupby('ageg', as_index = False) \
.agg(mean_income = ('income' ,'mean'))
'''
income의 결측치를 제거하고
ageg별로 그룹화, 인덱스 넘버 추가
mean_income이라는 수입의 평균을 변수에 저장
'''
ageg_income
```

```python
# 막대그래프
sns.barplot(data = ageg_income, x = 'ageg', y = 'mean_income')
```

```python
# 정렬하기(연령층별)
sns.barplot(data = ageg_income, x = 'ageg', y = 'mean_income', order = ['young', 'middle', 'old'])
```

## 성별 직업 빈도

종교 유무에 따른 이혼율표 만들기

```python
# etc는 제외대상이다.
# normalize = True를 이용해 백분위를 알 수 있따.
rel_div = welfare.query('marriage != "etc"') \
.groupby('religion', as_index = False) \
['marriage'] \
.value_counts(normalize = True)
rel_div
```

```python
# 그래프
# 백분율로 바꿈, 반올림 round
rel_div = rel_div.query('marriage == "divorce"') \
.assign(proportion = rel_div['proportion'] * 100) \
.round(1)
rel_div
```

```python
#막대그래프
sns.barplot(data = rel_div, x = 'religion', y = 'proportion')
```

연령대 및 종교 유무에 따른 이혼율 분석

```python
#연령대별 이혼율 표
age_div = welfare.query('marriage != "etc"') \
.groupby('ageg', as_index = False) \
['marriage'] \
.value_counts(normalize = True)
age_div
```

```python
# 연령대별 이혼율 그래프
# ageg에 young를 제외하고, 이혼만 추출, 소수점 첫째 자리까지 반올림
age_div = age_div.query('ageg != "young" & marriage == "divorce"') \
.assign(proportion = age_div['proportion'] * 100) \
.round(1)
age_div
```

```python
# 그래프
sns.barplot(data = age_div, x = "ageg", y = 'proportion')
```

```python
# 연령대 및 종교 유무에 따른 이혼율표 만들기

# 기타제외, young 제외
# ageg, religion 분리, marriage 추출, 비율추출
age_rel_div = welfare.query('marriage != "etc" & ageg != "young"') \
.groupby(['ageg', 'religion'], as_index = False) \
['marriage'] \
.value_counts(normalize = True)
age_rel_div
```

```python
# 그래프를 만들기 위한 표 만들기
# divorce 추출, 백분율로 만들고 반올림
age_rel_div = age_rel_div.query('marriage == "divorce"') \
.assign(proportion = age_rel_div['proportion'] * 100).round(1)
age_rel_div
```

```python
# 막대그래프
sns.barplot(data = age_rel_div, x = 'ageg', y = 'proportion', hue = 'religion')
```

## 지역별 연령대 비율

```python
welfare['code_region'].dtypes
```

```python
welfare['code_region'].describe()
```

```python
welfare['code_region'].value_counts()
```

1서울, 2수도권, 3부산경남울산, 4대구경북, 5대전충남, 6강원충북, 7광주전남전북제주도

```python
list_region = pd.DataFrame({'code_region' : [1, 2, 3, 4, 5, 6, 7],
                            'region' : ['서울', '수도권', '부산/경남/울산', '대구/경북', '대전/충남', '강원/충북', '광주/전남/전북/제주도']})
list_region
```

```python
# 변수들 추가
welfare = welfare.merge(list_region, how = 'left', on = 'code_region')
welfare[['code_region', 'region']].head()
```

```python
# 지역별 연령대 비율표 만들기
region_ageg = welfare.groupby('region', as_index = False)['ageg'].value_counts(normalize = True)
region_ageg
```

```python
# 그래프 만들기
region_ageg = region_ageg.assign(proportion = region_ageg['proportion'] * 100).round(1)
region_ageg
```

```python
sns.barplot(data = region_ageg, y = 'region', x = 'proportion', hue = 'ageg')
```

누적비율 막대 그래프 만들기
1. 피벗 : 행과 열을 회전해 구성을 바꾼다.  df.pivot()

2. 그래프 : 가로막대그래프 -> df.plot.barh() , 막대를 누적 -> stacked = True

```python
# 피벗
pivot_df = region_ageg[['region', 'ageg', 'proportion']].pivot(index = 'region', columns = 'ageg', values = 'proportion')
pivot_df
```

```python
#그래프
pivot_df.plot.barh(stacked = True)
```

```python
# 막대 정렬
order_df = pivot_df.sort_values('old')[['young', 'middle', 'old']]
order_df
```

```python
#그래프
order_df.plot.barh(stacked = True)
```
