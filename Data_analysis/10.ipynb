대통령 대선 출마 선언문이 담겨있는 speech_moon.txt

open()으로 열고 read()를 통해 불러온다.

```python
moon = open('/content/drive/MyDrive/Colab Notebooks/WE_IT_DATA_ANALYSIS/speech_moon.txt', encoding = 'UTF-8').read()
moon
```

```python
# 불필요한 문자 제거 -> re.sub()
import re
moon = re.sub('[^가-힣]', ' ', moon)
moon # [^가-힣] -> 한글이 아닌 모든 문자를 의미한다.
```

```python
# 명사 추출 konly.tag.hannanum() 의 nouns()
import konlpy
hannanum = konlpy.tag.Hannanum()

nouns = hannanum.nouns(moon)
nouns
```

```python
# df로 변환
import pandas as pd
df_word = pd.DataFrame({'word' : nouns})
df_word
```

```python
# 단어빈도표 만들기
# 글자 수 추가 -> pd.str.len()
df_word['count'] = df_word['word'].str.len()
df_word
```

```python
df_word = df_word.query('count >= 2')
df_word.sort_values('count')
```

```python
df_word = df_word.groupby('word', as_index = False).agg(n = ('word', 'count')).sort_values('n', ascending = False)
df_word
```

```python
top20 = df_word.head(20)
top20
```

## 기사 댓글 텍스트 마이닝

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/WE_IT_DATA_ANALYSIS/news_comment_BTS.csv', encoding = 'UTF-8')

df.info()
```

```python
# 불필요한 문자 제거 -> str.replace() : 데이터프레임 안에서 사용 가능
df['reply'] = df['reply'].str.replace('[^가-힣]', ' ', regex = True) #
df['reply'].head()
```

```python
# 명사 추출(꼬꼬마, Kkma 사용 -> 띄어쓰기 오류가 있는 문장에서도 잘 추출)
import konlpy
kkma = konlpy.tag.Kkma()
```

```python
# 명사 추출 - apply()활용 -> 리스트형식으로 변환
nouns = df['reply'].apply(kkma.nouns)
nouns
```

```python
# 한 행에 한 단어가 들어가도록 구성
nouns = nouns.explode()
nouns
```

```python
# 데이터 프레임 만들기
df_word = pd.DataFrame({'word' : nouns})

#글자 수 변수 추가
df_word['count'] = df_word['word'].str.len()

# 두 글자 이상 단어만 남기기
df_word = df_word.query('count >= 2')
df_word
```

```python
# 빈도표 만들기
df_word = df_word.groupby('word', as_index = False).agg(n = ('word', 'count')).sort_values('n', ascending = False)
df_word
```

```python
top20 = df_word.head(20)
top20
```

```python
# 가로 세로 크기 설정

# 막대 그래프
sns.barplot(data = top20, y = 'word', x = 'n')
```

워드클라우드

```python
# 데이터프레임을 딕셔너리로 변환
dic_word = df_word.set_index('word').to_dict()['n']
```

```python
wc = WordCloud(random_state = 1234,
               font_path = font,
               width = 400,
               height = 400,
               background_color = 'white',
               mask = img)
```

```python
img_wordcloud = wc.generate_from_frequencies(dic_word)

plt.figure(figsize = (10, 10))
plt.axis('off')
plt.imshow(img_wordcloud)
```
