## mpg 통합 연비 변수 만들기


mpg 데이터의 컬럼명 중

cty : 도시 연비

hwy : 고속도로 연비

```python
mpg.head()
```

```python
mpg['통합연비'] = (mpg['cty'] + mpg['hwy']) / 2
mpg.head()
```

### 조건문을 활용

#### 기준값 정하기

```python
mpg['통합연비'].describe() # 요약 통계량
```

```python
mpg['통합연비'].plot.hist() # 히스토그램
```

요약 통계량, 히스토그램을 통해 알 수 있는 결과로

통합연비의 평균, 중앙값은 20

통합연비는 20~22.5에서의 자동차모델이 많다

통합연비는 대부분 25이하며, 25를 넘기는 모델은 많지 않다.

통합연비가 20을 넘으면 합, 넘지 않으면 불로 분류한 변수 만들기

Numpy 패키지의 where() 이용 : 조건에 따라 서로 다른 값을 부여

np.where()

```python
import numpy as np
```

np.where(조건, 조건이 참일 때의 값, 조건이 거짓일 때의 값)

```python
mpg['test'] = np.where(mpg['통합연비']>=20, 'P', 'F')
mpg
```

```python
mpg['test'].value_counts() # 합,불의 빈도 확인
```

```python
count_test = mpg['test'].value_counts()
count_test.plot.bar() # 연비 합격 빈도 막대 그래프
#
```

축 이름 회전

```python
count_test.plot.bar(rot = 0)
```

### 중첩 조건문 활용

등급의 기준으로

A : 30이상

B : 20~29

C : 20미만 으로 정할 때,

```python
mpg['grade'] = np.where(mpg['통합연비'] >= 30, 'A',
                        np.where(mpg['통합연비'] >= 20, 'B', 'C'))
mpg.head()
```
