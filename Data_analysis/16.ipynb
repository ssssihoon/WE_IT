df.loc[] 이용

```python
df = pd.DataFrame({'var1' : [1, 2, 3],
                   'var2' : [4, 5, 6]})
df
# 딕셔너리형식으로 데이터프레임을 만든다.
```

```python
#인덱스 문자열로 지정
df = pd.DataFrame({'var1' : [1, 2, 3],
                   'var2' : [4, 5, 6]},
                  index = ['kim', 'lee', 'park'])
df
```

```python
# df.loc를 이용해 인덱스 문자열을 지정해 행을 추출
df.loc['kim']
```

```python
df.loc[:, 'var1']
```

df.loc[] 에서 인덱스 번호를 가지고도 추출이 가능하다.

만약 인덱스 번호가 없이 인덱스 문자열만 있다면 df.loc[]에서 인덱스 번호로 행을 추출하지 못한다. BUT-> 열은 추출 가능

연속된 행을 추출하려면 슬라이싱으로 추출하면 된다.

df.loc[x:y] : x행 이상 y행 이하 추출

df.iloc[x:y] : x행 이상 y행 미만 추출 -> 일반적인 파이썬의 방식 느낌

df.iloc[]의 경우 인덱스 문자열이 있어도 인덱스 번호로 행을 추출할 수 있다.

df.iloc[]의 경우 조건을 지정해 행을 추출할 수 없다.
