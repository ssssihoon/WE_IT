## lambda를 이용해 데이터 프레임명 줄이기

lambda를 이용하면 데이터 프레임명 대신 약어를 입력해 코드를 간결하게 작성할 수 있다.

```python
import pandas as pd
long_name = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/WE_IT_DATA_ANALYSIS/exam.csv')

long_name.assign(new = lambda x: x['math'] + x['english']).head()
#lambda x: <- 데이터 프레임명 자리에 x를 입력하겠다는 의미
#new라는 파생변수 추가
```

# 집단별로 요약

df.groupby() : 그룹화하는데 사용되며 그 이후에 연산 가능


df.agg() : 집계 연산을 한번에 처리할 때 사용

df.agg(할당할 변수 명 = (사용할 변수, 사용할 함수))

      df.agg(할당할 변수 명1 = (사용할 변수1, 사용할 함수1),

            할당할 변수 명2 = (사용할 변수2, 사용할 함수2),
       
             '''''')

```python
exam.agg(mean_math = ('math', 'mean'))
```

집단별 요약 통계량 구하기

```python
exam.groupby('nclass')\
    .agg(mean_math = ('math', 'mean'))
# nclass에 대해 집단별로 묶고 수학과목에 대한 평균값 출력
```

만약 변수를 인덱스로 바꾸지 않으려면

df.groupby('변수명', as_index = False)

## 집단별로 다시 집단 나누기

```python
import pandas as pd

mpg = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/WE_IT_DATA_ANALYSIS/mpg.csv")
mpg.head() # df
```

mpg 데이터프레임에서 제조 회사별 집단을 나눈 다음 다시 구동방식별로 나눠 도시 연비 평균 구하기

```python
mpg.groupby(['manufacturer', 'drv']) \
    .agg(mean_cty = ('cty', 'mean'))
# drv : 작동방식
# cty : 연비
```

이 중에서 한 제조사를 택 해 뽑는다면

```python
mpg.query('manufacturer == "audi"') \
    .groupby(['drv']) \
    .agg(n = ('drv', 'count'))
```

df.value_counts() 는 빈도를 구할 수 있고, 빈도 기준으로 내림차순 정렬을 한다. but 시리즈 자료 구조이기 때문에 query()를 적용할 수 없다.
